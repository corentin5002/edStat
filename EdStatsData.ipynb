```python

dfPopulation = dfEdData[dfEdData['Indicator Code'] == 'SP.POP.1524.TO.UN']
# Liste des colonnes à supprimer
colonnes_a_supprimer = [str(annee) for annee in range(1970, 1990)]

# Supprimer les colonnes
dfPopulation = dfPopulation.drop(colonnes_a_supprimer, axis=1)

colonnes_a_supprimer = [str(annee) for annee in dfPopulation.columns[4:]]

colonnes_a_supprimer = colonnes_a_supprimer[colonnes_a_supprimer.index('2017') + 1:]



dfPopulation = dfPopulation.drop(colonnes_a_supprimer, axis=1)

dfPopulation.head()
```

```python
# Plot for every country, the curve for the indicator 'SP.POP.1524.TO.UN' over the years with plotply

fig = go.Figure()

countries = dfEdData['Country Name'].unique()

for country in countries: 
    df_filtered = dfPopulation[dfPopulation['Country Name'] == country]

    fig.add_trace(go.Scatter(x=df_filtered.columns[4:], y=df_filtered.iloc[0, 4:], mode='lines', name=country))

fig.update_layout(
    title= "Évolution de la population qui suit des cours de 15 à 24 ans par pays",
    xaxis_title="Année",
    yaxis_title="Population",
    legend_title="Pays",
)
    
fig.show()
```
